## Binary Search Tree Implementation

```python
class Node: 
    def __init__(self,key): 
        self.left = None
        self.right = None
        self.val = key 
        
class BST:
    def __init__(self):
        self.root = None
        self.length = 0
        
    def insert(self,value):
        new_node = Node(value)
        
        if self.root is None: 
            self.root = new_node
            self.length +=1
        else:
            currentNode = self.root
            while(True):
#                 print(currentNode.val)
                if currentNode.val < new_node.val: 
                    if currentNode.right is None: 
                        currentNode.right = new_node
                        self.length +=1
                        return self.length
                    else: 
                        currentNode = currentNode.right
                else: 
                    if currentNode.left is None: 
                        currentNode.left = new_node 
                        self.length +=1
                        return self.length
                    else: 
                        currentNode = currentNode.left
                    
    def lookup(self,value):
        if self.root == None:
            return False
        currentNode = self.root;
        while(currentNode):
            if(value < currentNode.val):
                currentNode = currentNode.left
            elif(value > currentNode.val):
                currentNode = currentNode.right
            elif(value == currentNode.val):
                return currentNode.val
                
        return False 
    
    # inoder traversal of the BST 
    def inorder(self,root): 
        currentNode = root
        if root: 
            self.inorder(root.left) 
            print(root.val) 
            self.inorder(root.right)
            
    
            
    def breadthFirstSearch(self):
        """Breadth First Search"""
        currentNode = self.root
        queue = []
        bfsList = []
        queue.append(currentNode)
        
        while len(queue) > 0:
            currentNode = queue.pop(0)
#             print(currentNode.val)
            bfsList.append(currentNode.val)
            
            if currentNode.left:
                queue.append(currentNode.left)
            if currentNode.right:
                queue.append(currentNode.right)
                
        return bfsList
            
        
        
bst = BST()
bst.insert(1)
bst.insert(3)
bst.insert(10)
bst.insert(4)
bst.insert(5)
bst.insert(6)
print('Search {10} in bst',bst.lookup(10))
print('Search {21} in bst',bst.lookup(21))
print('\nInorder traverse of BST')
bst.inorder(bst.root)
print(bst.breadthFirstSearch.__doc__)
bst.breadthFirstSearch()
```

### Tree Traversals (Inorder, Preorder and Postorder) 

         9
      /    \
     4      20
    / \     / \
    1  6   15  170


#### Depth First Traversals:
1. Inorder (Left, Root, Right) : 1, 4, 6, 9, 15, 20, 170
2. Preorder (Root, Left, Right) : 9, 4, 1, 6, 20, 15, 170
3. Postorder (Left, Right, Root) : 1, 6, 4, 15, 170, 20, 9

#### Breadth First or Level Order Traversal 
> 9, 4, 20, 1, 6, 15, 170



```python
class Node: 
    def __init__(self,key): 
        self.left = None
        self.right = None
        self.val = key 
        
class BST:
    """BST"""
    def __init__(self):
        self.root = None
        self.length = 0
        
    def insert(self,value):
        new_node = Node(value)
        
        if self.root is None: 
            self.root = new_node
            self.length +=1
        else:
            currentNode = self.root
            while(True):
#                 print(currentNode.val)
                if currentNode.val < new_node.val:
#                     print('right-> ',value)
                    if currentNode.right is None: 
                        currentNode.right = new_node
                        self.length +=1
                        return self.length
                    else: 
                        currentNode = currentNode.right
                else: 
#                     print('left-> ',value)
                    if currentNode.left is None: 
                        currentNode.left = new_node 
                        self.length +=1
                        return self.length
                    else: 
                        currentNode = currentNode.left
    
    def breadthFirstSearch(self):
        """Breadth First Search"""
        currentNode = self.root
        queue = []
        bfsList = []
        queue.append(currentNode)
        
        while len(queue) > 0:
            currentNode = queue.pop(0)
#             print(currentNode.val)
            bfsList.append(currentNode.val)
            
            if currentNode.left:
                queue.append(currentNode.left)
            if currentNode.right:
                queue.append(currentNode.right)
                
        return bfsList
    def breadthFirstSearchRecursive(self,queue,bfsList):
        """Breadth First Search using Recursion"""
        if len(queue) == 0:
            return bfsList
        currentNode = queue.pop(0)
#         print(currentNode.val)
        bfsList.append(currentNode.val)
        if currentNode.left:
            queue.append(currentNode.left)
        if currentNode.right:
            queue.append(currentNode.right)
        
        return self.breadthFirstSearchRecursive(queue,bfsList)
    
    def DFSOLDWAYInorder(self,root): 
        """DFS -> 1. inoder traversal of the BST """
        currentNode = root
        if root: 
            self.DFSInorder(root.left) 
            print(root.val) 
            self.DFSInorder(root.right)
    def DFSInorder(self):
        """DFS Inorder uisng recursion -> LVR"""
        return traverseInOrder(self.root,[])
    def DFSPreorder(self):
        """DES Preorder uisng recursion -> VLR"""
        return traversePreOrder(self.root,[])
    def DFSPostorder(self):
        """DES Postorder uisng recursion -> LRV"""
        return traversePostOrder(self.root,[])

            
def traverseInOrder(node,InLst):
    """LVR -> Left Root RIght"""
#     print(node.val)
#     print(InLst)
    if node.left:
        traverseInOrder(node.left,InLst)
    InLst.append(node.val)
    if node.right:
        traverseInOrder(node.right,InLst)
    return InLst   

def traversePreOrder(node,PreLst):
    """VLR"""
#     print(node.val)
#     print(PreLst)
    PreLst.append(node.val)
    if node.left:
        traversePreOrder(node.left,PreLst)
    if node.right:
        traversePreOrder(node.right,PreLst)
    return PreLst

def traversePostOrder(node,PostLst):
    """LRV"""
#     print(node.val)
#     print(PostLst)
    if node.left:
        traversePostOrder(node.left,PostLst)
    if node.right:
        traversePostOrder(node.right,PostLst)
    PostLst.append(node.val)
    
    return PostLst
        
        
bst = BST()
bst.insert(9)
bst.insert(4)
bst.insert(6)
bst.insert(20)
bst.insert(170)
bst.insert(15)
bst.insert(1)
print(bst.breadthFirstSearch.__doc__)
print(bst.breadthFirstSearch())
print(bst.breadthFirstSearchRecursive.__doc__)
print(bst.breadthFirstSearchRecursive([bst.root],[]))
print(bst.DFSInorder.__doc__)
print(bst.DFSInorder())
print(bst.DFSPreorder.__doc__)
print(bst.DFSPreorder())
print(bst.DFSPostorder.__doc__)
print(bst.DFSPostorder())
```
